```python
import pandas as pd

# Lendo o arquivo CSV
df = pd.read_csv('charts.csv', low_memory=False)
```

```python
print("\nAparições de 'Coldplay' no campo 'artist':")

pd.set_option('display.max_rows', None)  # Isso vai mostrar todas as linhas sem truncar

# Filtrando e mostrando as linhas de uma música
coldplay_tracks = df[df['artist']=='Coldplay']
coldplay_tracks = coldplay_tracks[coldplay_tracks['region']=='Global']
print(coldplay_tracks.iloc[:,:4])
```

```python
# Filtrando e mostrando as linhas de uma música
top = df[df['date']=='2020-01-01']
top = top[top['region']=='Global']
top = top[top['chart']=='top200']
top = top[top['rank']==1]

print(top)
```
